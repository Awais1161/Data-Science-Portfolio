# Multicolinearity (VIF)

```python
df_num=df.select_dtypes(include='number')
```

```python
X = df_num

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)

calc_vif(X)
```
